```python
# ✅ Step 1: Install dependencies (FREE models only)
!pip install -q langchain faiss-cpu sentence-transformers pypdf langchain-community transformers

# ✅ Step 2: Import modules
import os
from google.colab import files
from langchain_community.document_loaders import PyPDFLoader
from langchain.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS
from langchain.chains import RetrievalQA
from langchain.llms import HuggingFacePipeline
from transformers import pipeline

# ✅ Step 3: Upload your menu PDF
print("📄 Please upload your menu PDF file...")
uploaded = files.upload()
pdf_file = next(iter(uploaded))

# ✅ Step 4: Load and split the PDF
loader = PyPDFLoader(pdf_file)
pages = loader.load_and_split()

if not pages:
    raise ValueError("❌ No text found in PDF. Use a searchable (not scanned) menu.")

# ✅ Step 5: Embed text using HuggingFace model
print("🔍 Creating vector database...")
embeddings = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
db = FAISS.from_documents(pages, embeddings)

# ✅ Step 6: Load a FREE small chatbot model (FLAN-T5)
print("🤖 Loading chatbot model (flan-t5-base)...")
qa_pipeline = pipeline(
    "text2text-generation",
    model="google/flan-t5-base",
    tokenizer="google/flan-t5-base",
    max_length=256,
    temperature=0.0
)

llm = HuggingFacePipeline(pipeline=qa_pipeline)

# ✅ Step 7: Build the chatbot QA chain
qa_chain = RetrievalQA.from_chain_type(
    llm=llm,
    chain_type="stuff",
    retriever=db.as_retriever(search_kwargs={"k": 3}),
    return_source_documents=False
)

# ✅ Step 8: Chat with the bot
print("\n✅ 🍕 Menu Chatbot is ready!")
print("Ask anything about the menu (items, prices, availability, suggestions, etc.)")
print("Type 'exit' or 'quit' to end the conversation.\n")

while True:
    query = input("Customer: ")#
    if query.lower() in ['exit', 'quit']:
        print("👋 Goodbye!")
        break
    try:
        response = qa_chain.invoke({"query": query})
        print(f"Bot: {response['result']}\n")
    except Exception as e:
        print("⚠️ Error:", str(e))
```
